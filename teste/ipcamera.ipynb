```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
cam = cv.VideoCapture("http://192.168.2.101:8080/video")
```

```python
ret, frame = cam.read()

plt.imshow(frame)
```

```python
frame
```

```python
tempo1 = cv.getTickCount()


test = frame.flatten()

test.tolist()

height = frame.shape[0]
width = frame.shape[1]

tempo2 = cv.getTickCount()

print((tempo2-tempo1)/cv.getTickFrequency())
```

```python

img = test
img = np.array(np.split(img, height))
img = np.array(np.split(img, width, axis=1))

img = np.rot90(img)
```
